## Integração com banco de dados SQL

* Enviando query para o banco
* Consumindo tabela do banco
* Enviando dados para o banco

```python
import sqlalchemy
import pandas as pd

con = sqlalchemy.create_engine("sqlite:///data/banco.db")

con.table_names()
```

```python
query = '''

SELECT *

FROM tb_votacao

WHERE ano_eleicao = 2018

'''
```

```python
df_2018 = pd.read_sql_query( query, con )
df_2018
```

```python
query_grande = '''

SELECT t1.sigla_partido,
       sum(t2.valor) / count( distinct t1.numero_sequencial) as avg_valor

from tb_votacao as t1

left join tb_bens as t2
on t1.numero_sequencial = t2.numero_sequencial

where t1.ano_eleicao = 2018

group by t1.sigla_partido

'''

df_candidato_bens = pd.read_sql_query(query_grande, con)
df_candidato_bens
```

```python
df_candidato_bens.to_csv("Sumarios_partido_bens.csv", index=False)
df_candidato_bens.to_excel("Sumarios_partido_bens.xlsx", index=False)
```

```python
df_candidato_bens.to_sql("tb_bens_sumario", con, index=False)
```
